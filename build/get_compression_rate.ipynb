#PLOT C RATE FOR MAXIMUM LABEL SIZE


```python
path = "/home/dtkhuat/results/random_graphs/var_edge_len/"
str_len = 4
increment = 4
c_size = []
original = []
ir_c = []
ie_c = []
date = '06_06_2024.csv'
while str_len < 65:
    temp = []
    temp2 = []
    temp3 = []
    with open(path+str(str_len)+"/"+date, 'r') as csv_file:
        csvreader = csv.reader(csv_file)
        for row in csvreader:
            temp.append(float(row[1]))
            temp2.append(float(row[2]))
            temp3.append(float(row[3]))
    original.append(temp)
    ir_c.append(temp2)
    ie_c.append(temp3)
    str_len = str_len + increment

fig = plt.figure(figsize =(5, 3))
ax = fig.add_axes([0, 0, 1, 1])
plt.xlabel("Maximum Length of Node and Edge Labels")
plt.ylabel("Compression ratio")
#bp = ax.stem(timings,showfliers=False,showmeans=True)
x = []
y = []

for i in range(0,16):
    y.append(statistics.mean(ir_c[i])/statistics.mean(original[i]))
    x.append(((i+1)*4)-0.5)
bp = ax.bar(x, y, linewidth=1, capsize=6, label="ITR on RDF graphs")



x2 = []
y2 = []

for i in range(0,16):
    y2.append(statistics.mean(ie_c[i])/statistics.mean(original[i]))
    x2.append(((i+1)*4+0.5))
bp = ax.bar(x2, y2, linewidth=1, capsize=6, label="ITR on Graphs with indexed Edges")

plt.legend(loc="upper right")
plt.xticks(np.arange(0, 65,  step=4))
plt.yticks(np.arange(0, 1.1, step=0.1))
plt.savefig("/home/dtkhuat/plots/test/random_var_edge_len3.pdf", format="pdf", bbox_inches="tight")
plt.show()

plt.close()

print(np.arange(0, 64, step=4))

'''
fig = plt.figure(figsize =(5, 5))
ax = fig.add_axes([0, 0, 1, 1])
plt.xlabel("Maximum label size of node and edges")
plt.ylabel("Compression ratio")
x = []
y = []

for i in range(0,16):
    y.append(statistics.mean(ie_c[i])/statistics.mean(original[i]))
    x.append((i+1)*4)
bp = ax.bar(x, y, linewidth=2, capsize=6)
plt.savefig("/home/dtkhuat/plots/test/ie_var_edge_len221313.pdf", format="pdf", bbox_inches="tight")
plt.show()
plt.close()
'''
```
